```python

df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])

df['Returns'] = df['Returns'].fillna(0).astype(int)

df.dropna(subset=['Customer Name'], inplace=True)

customer_purchase_counts = df.groupby('Customer Name')['Purchase Date'].nunique()


customers_bought_on_different_days = customer_purchase_counts[customer_purchase_counts > 1]

return_counts = df[df['Customer Name'].isin(customers_bought_on_different_days.index)].groupby('Customer Name')['Returns'].sum()


customer_counts_df = pd.DataFrame({
    'Customer Name': return_counts.index,
    'Purchase Counts': customers_bought_on_different_days.values,
    'Return Counts': return_counts.values
})


customer_counts_df

```
